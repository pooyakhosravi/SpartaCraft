### Normal Q Learning

```python
env = gym.make('FrozenLake-v0')
QTable = Agent.QTable(env)
normal_agent = Agent.QLearning(env, QTable)
normal_agent.env.render()
j_list, r_list = normal_agent.train(10000)

print("Final Q-Table Values")
normal_agent.render()

result = sum(r_list)/len(r_list)
print(f"Score over time: {result}")
Utils.graph_results(j_list, r_list)
```

```python
r_list = normal_agent.test(10000)
result = sum(r_list)/len(r_list)
print(f"Score over time: {result}")
```
